```python
import numpy as np
jarvis_energy = np.array([30,25,35,40,28,32,36])

# Calculate total energy consumption
total_energy = np.sum(jarvis_energy)

# Calculate average daily energy consumption
average_daily_energy = np.mean(jarvis_energy)

print("Total energy consumption : ",total_energy)
print("Average daily consumption : ",average_daily_energy)
```

```python
import matplotlib.pyplot as plt

#Adventure data
adventures = ["Battle of New York","Rescue Mission","Infiltration","Technology Breakthrough","Defeating Villain"]
success_rates = [90,75,80,95,70] #Success rates for each adventure

#Create a line plot
plt.figure(figsize =(8,5))
plt.plot(adventures,success_rates,marker ='o',linestyle ='-',color ='cyan')
plt.title("Iron man's Heoric Journey")
plt.xlabel("Adventures")
plt.ylabel("Success Rate (%)")
plt.xticks(rotation = 45)
plt.grid(True)

#Create a pie chart
labels = adventures
plt.figure(figsize =(6,6))
plt.pie(success_rates,labels = labels,autopct = '%1.1f%%',startangle = 140)
plt.title("Types of mission")
```
